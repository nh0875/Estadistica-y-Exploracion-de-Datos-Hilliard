```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

vinos = pd.read_csv('/content/caracteristicas de vinos.csv')
print(vinos.to_string())
```

```python
vinos.info()
vinos.head()
```

```python
vinos_var = vinos.drop(['Vino'], axis=1)
vinos_var.describe()
```

```python
vinos_normalized = (vinos_var-vinos_var.min())/(vinos_var.max()-vinos_var.min())
vinos_normalized
```

```python
vinos_normalized.describe()
```

```python
wcss = []
for t in range(1,11):
  kmeans = KMeans(n_clusters = t, random_state = 0)
  kmeans.fit(vinos_normalized)
  wcss.append(kmeans.inertia_)
plt.plot(range(1,11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
clustering = KMeans(n_clusters=3, max_iter = 300)
clustering.fit(vinos_normalized)
```

```python
vinos['Kmeans_Clusters'] = clustering.labels_
vinos.head()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
vinos_pca = pca.fit_transform(vinos_normalized)
vinos_pca_df = pd.DataFrame(data = vinos_pca, columns = ['Component_1', 'Component_2']) # Modificación aquí
vinos_pca_names = pd.concat([vinos_pca_df, vinos[['Kmeans_Clusters']]], axis = 1)
vinos_pca_names
```

```python
fig = plt.figure(figsize = (6,6))

ax = fig.add_subplot(1,1,1)
ax.set_xlabel('Component 1', fontsize = 15)
ax.set_ylabel('Component 2', fontsize = 15)
ax.set_title('Principal Component', fontsize = 20)

color_theme = np.array(['red', 'yellow', 'pink'])
ax.scatter(x = vinos_pca_names.Component_1, y = vinos_pca_names.Component_2,
           c = color_theme[vinos_pca_names.Kmeans_Clusters], s = 50)
plt.show()
```

# **Grafico Interactivo**
Pregunta a la que responde: ¿Cómo están distribuidos los diferentes grupos de vinos en el espacio reducido a dos componentes principales?

```python
import plotly.express as px
fig_interactive = px.scatter(vinos_pca_names, x='Component_1', y='Component_2',
                             color=vinos_pca_names['Kmeans_Clusters'].astype(str),
                             title="Distribución interactiva de clusters en el plano PCA",
                             labels={'Component_1': 'Componente Principal 1', 'Component_2': 'Componente Principal 2'},
                             color_discrete_sequence=['red', 'yellow', 'blue'])
fig_interactive.show()
```

# **Grafico estatico**

Responde a la pregunta: ¿Cuál es el nivel promedio de alcohol en cada uno de los clusters de vinos?

```python
import matplotlib.pyplot as plt

alcohol_mean_by_cluster = vinos.groupby('Kmeans_Clusters')['Alcohol'].mean()
plt.figure(figsize=(8, 5))
alcohol_mean_by_cluster.plot(kind='bar', color=['red', 'yellow', 'pink'])
plt.title("Nivel promedio de alcohol por cluster")
plt.xlabel("Clusters de Vino")
plt.ylabel("Nivel Promedio de Alcohol")
plt.xticks(rotation=0)
plt.show()
```
